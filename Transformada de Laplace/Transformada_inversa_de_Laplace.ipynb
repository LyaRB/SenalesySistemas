```python
import sympy as sym
%matplotlib inline
```

```python
s = sym.symbols('s', complex=True)
t = sym.symbols('t', real=True)

X = (1/((s+1)*(s+2)**2))
X
```

```python
polos = sym.roots(sym.denom(X))
polos
```

```python
x1 = sym.inverse_laplace_transform(X, s, t)
x1
```

```python
sym.plot(x1, (t, 0, 10), xlabel = '$t$', ylabel = '$x(t)$');
```
